```python
import pandas as pd
from sklearn.model_selection import train_test_split
from emails import email_contents #another email program to extract contents and labels for email
```

```python
labels=[]
contents=[]
emailLabels=[]

#splitting labels and contens for sms
with open('SMSSpamCollection','r') as f:
	filer=f.readlines()
	for line in filer:
		labels.append(line[:4].strip()) 
		contents.append(line[4:].strip())

#splitting labels and contnets of mail 
with open('E:\python projects\Spam or Ham\CSDMC2010_SPAM\SPAMTrain.label' ,'r') as e:	
	email=e.readlines()
	for e in email:
		emailLabels.append(e.split(" ")[0].strip())

#changing labels "1" to ham and "0" to Spam
for i in range(len(emailLabels)):
	if emailLabels[i]=='1':
		emailLabels[i]='ham'
	else:
		emailLabels[i]='spam'
# print(emailLabels)		
```

```python
email_content,email_test_cont=email_contents() #calling email_contents() to get train and test emails from emails.py
```

```python
#splitting sms dataset: 80% for training and 20% for testing 
labels_train,labels_test,contents_train,contents_test=train_test_split(labels,contents,test_size=0.2,random_state=42)
#splitting email dataset: 80% for training and 20% for testing 
email_label_train,email_label_test,email_content_train,email_content_test=train_test_split(emailLabels,email_content,test_size=0.2,random_state=42)
```

```python
#combining sms and email Dataset

labels_train.extend(email_label_train)
labels_test.extend(email_label_test)
contents_train.extend(email_content_train)
contents_test.extend(email_content_test)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count=CountVectorizer(stop_words="english")  # this helps to create bag of words

sms_transform=count.fit_transform(contents_train) #Extract token counts out of raw text documents using the vocabulary fitted with fit 
sms_test=count.transform(contents_test)
count.vocabulary_
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
features=TfidfTransformer()

feat=features.fit_transform(sms_transform)
contents1=features.transform(sms_test)
print (feat.shape)
print (feat.toarray())
```

```python
from sklearn import svm
clf=svm.SVC(C=10000.0,kernel="rbf",gamma=0.6) #Create SVM object with necessary parameters
clf.fit(feat,labels_train) #fit the training labels and contents
pred=clf.predict(contents1) #now predict the contents of test set
print (pred)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(labels_test,pred)  #find accuracy according to prediction

print (acc)
```
